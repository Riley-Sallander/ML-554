```raw
"""
Returns plural of a noun
  
Parameters
----------
gift: str
  A noun
    
Return
------
str
  Plural version
"""

def pluralize_gift(gift):
  if"oo" in gift :
   gift=gift.replace("oo","ee")
  elif gift.endswith("y"):
    gift = gift[:-1]+"ies"
  else:
    gift=gift+"s"

  return gift


pluralize_gift("goose")
# Will work if your function is vectorized! 
xmas['Gift.Item'].apply(pluralize_gift)
```

### Function 2

```raw

import inflect
p = inflect.engine()

def make_phrase(num, num_word, item, verb, adjective, location):
  """
  returns a full phrase based on the following inputs

Parameters 
-------
num: numerical value representing the day and number of gifts
num_word: the string form of num: not actually an input but at this point im scared I will break my code
item: string input of what gift was given
verb: string what the gift is doing 
adjective: string descriptor of gift 
location: string where the gift is


Return
-------
  string: returns the fully compiled phrase
    
  """
  
  # Step 1: Replace NAs (None) with blank strings
  verb = verb if pd.notna(verb) else ""
  adjective = adjective if pd.notna(adjective) else ""
  location = location if pd.notna(location) else ""

  # Step 2: Pluralize item if num > 1 and convert number to words
  if num > 1: 
    item = pluralize_gift(item)
    num_word = p.number_to_words(num)
  
  # Step 3/4:
  if num == 1: 
    if item[0].lower() in "aeiou": 
      num_word = "an"
    else:
      num_word = "a"

  # Create phrase
  parts = [num_word] 
  if adjective:
      parts.append(adjective)
  parts.append(item)
  if verb:
      parts.append(verb)

  phrase = ' '.join(parts).strip()
  
  # Add location if provided
  if location:
    phrase += f" {location}"
  return phrase.strip()


xmas['Full.Phrase'] = xmas.apply(lambda row: make_phrase(
    row['Day'], "", row['Gift.Item'], row['Verb'], row['Adjective'], row['Location']
), axis=1)

print(xmas['Full.Phrase'])
```

### Function 3

```raw
"""
  returns the full song based on the following inputs

Parameters 
-------
df: dataframe of information that is used in function create phrase as well as sing day

day: numerical value that indicates where to start the song from

Return
-------

full_song: string the full song from the starting point of the given day

 
"""



def sing_day(df, day):

    df['Full.Phrase'] = df.apply(lambda row: make_phrase(
    row['Day'], "", row['Gift.Item'], row['Verb'], row['Adjective'], row['Location']
), axis=1)

    full_song = "" #updates each day
    
    # Loop from the given day down to 1 to get verses in reverse order
    for current_day in range(day, 0, -1):
        nthDay = p.ordinal(p.number_to_words(current_day))
        verse = f"On the {nthDay} day of Christmas, my true love sent to me:\n"
        
        # Loop through all days up to the current day to create the verse
        for i in range(current_day, 0, -1):
            phrase = df.iloc[i - 1]["Full.Phrase"]
            
            # Special condition for the last line on day 1
            if i == 1 and current_day > 1:
                verse += f"and {phrase}.\n"
            else:
                verse += f"{phrase},\n"
        
        # Append the verse for this day to the full song
        full_song += verse.strip() + "\n\n"
    
    return full_song.strip()


print(sing_day(xmas,2))
```
