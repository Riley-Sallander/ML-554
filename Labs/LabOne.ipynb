```python

import pandas as pd 
dfQ1 = pd.read_csv("/Users/rileysallander/Desktop/ML/ML554/Data/q1data.csv")
dfQ1 = dfQ1.dropna()
dfQ1.head()
```

#### 3)

```python
from plotnine import geom_point, ggplot, aes, scale_x_log10,scale_y_continuous,theme_minimal,scale_size_area,scale_fill_manual,theme,annotate, element_blank,element_line,labs,scale_x_continuous,scale_y_log10
#basic graph aes assignments
(ggplot(dfQ1,
aes(
  x = "income",
  y = "life_exp",
  fill = "four_regions",
  size = "population"
))

# Scale the graph and sizes of bubbles along with colors
+ geom_point(color='black', stroke = 0.1)
+ scale_x_log10(breaks=[500,1000,2000,4000,8000,16000,32000,64000],limits=(500, 64000))
+ scale_y_continuous(breaks=[10,20,30,40,50,60,70,80,90],limits=(20, 90))
+ scale_size_area(max_size=15)
+ scale_fill_manual(values={
        'asia': '#ff5872',
        'europe': '#ffe700',
        'americas': '#7feb00',
        'africa': '#00d5e9',
    })

# Add themes and gridlines 
+ theme_minimal()
+ theme(figure_size=(8, 4))
+ theme(panel_grid_minor=element_blank(),  
        panel_grid_major=element_line(size=0.5),
        axis_line=element_line(size=0.5, color='grey',alpha = 0.8))

# Add axis lables and background 2010
+ annotate('text', x=6000, y=50, label='2010', size=100, color='grey', alpha=0.3, angle=0)
+ annotate('text', x=29000, y=20, label='per person(GDP/capita,PPPS infaltion-adjusted)', size=8, color='grey', alpha=0.9, angle=0)
+ annotate('text', x=500, y=90, label='year', size=8, color='grey', alpha=0.9, angle=90)

# Label the Axis
+ labs( 
  x='Income',  
  y='Life Expectancy'
  )
)
```

#### 4) 
<p>For an alternative graph I chose to create a density plot comparing the life expectancy of each region more specifically.
```python
from plotnine import geom_density

(ggplot(dfQ1, aes(x='life_exp', fill='four_regions')) 
+geom_density(alpha=0.6)
+labs(title='Life Expectancy by Region', y='Life Expectancy')
+theme_minimal()  # Clean theme for better visualization

+ scale_x_continuous(breaks=[40,50,60,70,80,90,100],limits=(30,100))
+ scale_fill_manual(values={
        'asia': '#ff5872',
        'europe': '#ffe700',
        'americas': '#7feb00',
        'africa': '#00d5e9',
    })


+ theme(figure_size=(8, 4))
+ theme(panel_grid_minor=element_blank(),  
        panel_grid_major=element_line(size=0.5),
        axis_line=element_line(size=0.5, color='grey',alpha = 0.8))

)
```

```python
dfQ2 = pd.read_csv("/Users/rileysallander/Desktop/ML/ML554/Data/q2data.csv")

dfQ2=dfQ2.dropna()

## converted thousand denoted with k to floats (consulted chatgpt on this one)
def convert_energy(value):
    if 'k' in value:
        return float(value.replace('k', '').strip()) * 1000
    else:
        return float(value)
        
dfQ2['energy'] = dfQ2['energy'].apply(convert_energy)

dfQ2.head()

dfQ2
```

#### 3)

```python


(ggplot(dfQ2,
aes(
  x = "exports",
  y = "imports",
  fill = "four_regions",
  size = "energy"
))

# Scale the graph and sizes of bubbles along with colors
+ geom_point(color='black', stroke = 0.1)
+ scale_x_continuous(breaks=[20,40,60,80,100,120,140,160,180,200,220],limits=(0,220))
+ scale_y_continuous(breaks=[0,50,100,150,200,250,300,350,400],limits=(0, 400))
+ scale_size_area(max_size=15)
+ scale_fill_manual(values={
        'asia': '#ff5872',
        'europe': '#ffe700',
        'americas': '#7feb00',
        'africa': '#00d5e9',
    })
# Add themes and gridlines 
+ theme_minimal()
+ theme(figure_size=(8, 4))
+ theme(panel_grid_minor=element_blank(),  
        panel_grid_major=element_line(size=0.5),
        axis_line=element_line(size=0.5, color='grey',alpha = 0.8))

# Add axis lables and background 2010
+ annotate('text', x=120, y=200, label='1997', size=100, color='grey', alpha=0.3, angle=0)

# Label the Axis
+ labs( 
  x='Exports (\% of GDP)',  
  y='Imports (\% of GDP)'
  )
)
```

#### 4)

```python
from plotnine import geom_boxplot

(ggplot(dfQ2,
        aes(x="four_regions", y="exports", fill="four_regions"))
+ geom_boxplot()
+ scale_fill_manual(values={
        'asia': '#ff5872',
        'europe': '#ffe700',
        'americas': '#7feb00',
        'africa': '#00d5e9',
    })

+ theme_minimal()
+ theme(figure_size=(8, 4))
+ theme(panel_grid_minor=element_blank(),  
        panel_grid_major=element_line(size=0.5),
        axis_line=element_line(size=0.5, color='grey',alpha = 0.8))

+ labs( 
  x='Region',  
  y='Exports (\% of GDP)'
  )


)
```

#### 4) 
<p>For this alternative data vizualization I created a heat map representing internet using % on the x axis and GDP on the y axis, while the heat was the count of occurance within each bin. The majority of the density is found in the lower left corner likely as small countries have smaller gdp and less people using the internet. This is not a great representation of the data as some of it is lost, like region and income, while not providing better insight. From the previous bubble chart you can see density through the bubble clustering while still retaining the lost information.<p>

```python
from plotnine import geom_bin2d, scale_fill_gradient

(ggplot(dfQ3,
        aes(x="internet_users", y="gdp"))
 + geom_bin2d(bins=40)
+ scale_x_continuous(breaks=[10,20,30,40,50,60,70,80,90],limits=(0,100))
+ scale_y_log10(breaks=[200,500,1000,2000,5000,10000,20000,50000,100000],limits=(200, 100000))

 + scale_fill_gradient(low="blue", high="red")

 # Add themes and gridlines 
+ theme(figure_size=(8, 4))
+ theme(panel_grid_minor=element_blank(),  
        panel_grid_major=element_line(size=0.5),
        axis_line=element_line(size=0.5, color='grey',alpha = 0.8))

# Label the Axis
+ labs( 
  x='Individuals Using The Internet',  
  y='GDP/capita'
  )
)
```
